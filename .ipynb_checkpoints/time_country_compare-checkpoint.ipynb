```python
### Import the necessary libraries and dependencies.
# Import libraries and dependencies
import pandas as pd
from pathlib import Path

%matplotlib inline
```

```python
### Read Australia data CSV in as DataFrame
# Set the path
csvpath = Path("ufo_df_au.csv")

# Read in the CSV as a DataFrame
au_df = pd.read_csv(
    csvpath, index_col="datetime", infer_datetime_format=True, parse_dates=True
)
au_df = au_df.sort_index()
au_df.head(5)
```

```python
### Read in US data CSV in as DataFrame
# Set the path
us_csvpath = Path("ufo_df_us.csv")

# Read in the CSV as a DataFrame
us_df = pd.read_csv(
    us_csvpath, index_col="datetime", infer_datetime_format=True, parse_dates=True
)
us_df = us_df.sort_index()
us_df.head(5)
```

```python
### Read in India data CSV as DataFrame
# Set the path
in_csvpath = Path("ufo_df_in.csv")

# Read in the CSV as a DataFrame
in_df = pd.read_csv(
    in_csvpath, index_col="datetime", infer_datetime_format=True, parse_dates=True
)
in_df = in_df.sort_index()
in_df.head(5)
```

```python
### Sightings per weekday dataframe - Australia
# Create dataframe with number of sightings per day of the week
au_sightings_day  = au_df[["day","sightings"]]
au_sightings_per_day = au_sightings_day.groupby("day").sum().reset_index()
au_sightings_per_day
```

```python
# Create a new index which corresponds with order of days of the week
new_index = [5,1,6,0,4,2,3]
au_sightings_per_day["new index"] = new_index
au_sightings_per_day = au_sightings_per_day.set_index(["new index"])

# Put days of the week in order by new index
au_sightings_per_day = au_sightings_per_day.sort_index(ascending=True)

au_sightings_per_day
```

```python
# Get number of sightings per weekday as a percentage
au_sightings_per_day['percent'] = (au_sightings_per_day['sightings'] / 
                              au_sightings_per_day['sightings'].sum()) * 100

au_sightings_per_day
```

```python
# Plot percentage of sightings per weekday on a bar chart
au_sightings_per_day.plot.bar('day','percent', rot=90, color="red", xlabel = "Day of Week", ylabel="Percent", title = "Australia UFO Sightings by Day of Week")
```
